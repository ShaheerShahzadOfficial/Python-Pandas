# ***Subsetting Data***

```python
df.shape
```

```python
df.columns
```

```python
df = df[['title', 'description', 'publishTime', 'duration_seconds',
       'viewCount', 'likeCount', 'commentCount']]
```

```python
df.head()
```

```python
df.shape
```

```python
#  Subsetting Rows

# subsetting using loc
df_subset1 = df.loc[df['viewCount'] > 1_000_000]
```

```python
# subsetting using query
df_subset2 =  df.query('viewCount > 1_000_000')
```

```python
df = df.loc[~df['viewCount'].isna()]
df = df.loc[~df['likeCount'].isna()]
```

# ***Casting dtypes***

```python
df['viewCount'] = df['viewCount'].astype('int')
df['likeCount']  = df['likeCount'].astype('int')
df['publishTime'] = pd.to_datetime(df['publishTime'])
```

```python
df.dtypes
```

# ***Creating new Column***

```python
df['like_to_view_ratio'] = df['likeCount'] / df['viewCount']
```

```python
df.head()
```

# ***Adding new Row***

```python
df_to_append = df.tail(1)

pd.concat([df,df_to_append])
```
